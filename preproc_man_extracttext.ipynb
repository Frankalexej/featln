```python
import pandas as pd
import os

from ssd_paths import *
```

```python
# Open the text file and read lines
file_path = os.path.join(ssd_aishell_path, "content.txt")  # Replace with the actual file path
with open(file_path, 'r', encoding='utf-8') as file:
    lines = file.readlines()

# Initialize lists to store data
rec_list = []
chinese_list = []
pinyin_list = []

# Process each line
for line in lines:
    # Split the line into "rec" and the sentence part
    parts = line.strip().split('\t')
    rec = parts[0]
    sentence = parts[1]

    # Split the sentence into Chinese characters and Pinyin
    sentence_parts = sentence.split()
    chinese_chars = ''.join(sentence_parts[0::2])
    pinyin = ' '.join(sentence_parts[1::2])

    # Append data to lists
    rec_list.append(rec)
    chinese_list.append(chinese_chars)
    pinyin_list.append(pinyin)

# Create a DataFrame
df = pd.DataFrame({
    'rec': rec_list,
    'chinese': chinese_list,
    'pinyin': pinyin_list
})
df
```

```python
df.to_csv(os.path.join(ssd_aishell_path, "aishell_transcript.csv"), index=False)
```
