## Init Model

```python
import torch.nn as nn
from model_config import *

class ResBlock(nn.Module):
    def __init__(self, n_chans):
        super(ResBlock, self).__init__()
        self.lin1 = nn.Linear(n_chans, n_chans)
        self.lin2 = nn.Linear(n_chans, n_chans)
        # self.batch_norm = nn.BatchNorm1d(num_features=n_chans)  # <5>
        self.batch_norm = nn.LayerNorm(n_chans)
        self.dropout = nn.Dropout(p=DROPOUT)
        self.relu = nn.ReLU()
        # self.relu = nn.ReLU()
        # self.relu = nn.Tanh()

    def forward(self, x):
        out = self.batch_norm(x)
        out = self.relu(out)
        out = self.lin1(out)
        out = self.batch_norm(out)
        out = self.relu(out)
        out = self.lin2(out)
        out = self.dropout(out)
        out = out + x
        return out

class LinPack(nn.Module):
    def __init__(self, n_in, n_out):
        super(LinPack, self).__init__()
        self.lin = nn.Linear(n_in, n_out)
        self.relu = nn.ReLU()
        # self.relu = nn.ReLU()
        # self.relu = nn.Tanh()
        # self.batch_norm = nn.BatchNorm1d(num_features=n_out)
        self.batch_norm = nn.LayerNorm(n_out)
        self.dropout = nn.Dropout(p=DROPOUT)

    def forward(self, x):
        x = self.lin(x)
        x = self.relu(x)
        x = self.batch_norm(x)
        x = self.dropout(x)
        return x


class ResAE(nn.Module):
    def __init__(self, input_dim=INPUT_DIM, inter_dim1=INTER_DIM_1, inter_dim2=INTER_DIM_2, inter_dim3=INTER_DIM_3, latent_dim=LATENT_DIM, output_dim=OUTPUT_DIM):
        super(ResAE, self).__init__()

        self.encoder = nn.Sequential(
            LinPack(input_dim, inter_dim1), 
            # LinPack(inter_dim1, inter_dim3), 
            ResBlock(inter_dim1), 
            # ResBlock(inter_dim1), 
            nn.Linear(inter_dim1, latent_dim * 2), 
            nn.Sigmoid()
        )

        self.decoder =  nn.Sequential(
            LinPack(latent_dim, inter_dim1), 
            # LinPack(inter_dim3, inter_dim1),
            ResBlock(inter_dim1), 
            # ResBlock(inter_dim1), 
            nn.Linear(inter_dim1, output_dim),
            # nn.Sigmoid(),
        )

        # initialize the weights
        self.encoder.apply(self.init_weights)
        self.decoder.apply(self.init_weights)
        
    def init_weights(self, m):
        if isinstance(m, nn.Linear):
            nn.init.xavier_uniform_(m.weight)
            # nn.init.kaiming_normal_(m.weight, mode='fan_out', nonlinearity='relu')
            m.bias.data.fill_(0.01)

    def reparameterise(self, mu, logvar):
        if self.training:
            # std = logvar.mul(0.5).exp_()
            # eps = std.data.new(std.size()).normal_()
            # res = eps.mul(std).add_(mu)
            epsilon = torch.randn_like(mu)
            res = mu + epsilon * torch.exp(logvar / 2)
            return res
        else:
            return mu

    def forward(self, x):
        org_size = x.size()
        y_size = (org_size[0], org_size[1], org_size[2] // 3)
        batch = org_size[0]
        x = x.view(batch, -1)

        h = self.encoder(x)
        mu, logvar = h.chunk(2, dim=1)
        z = self.reparameterise(mu, logvar)
        recon_x = self.decoder(z).view(size=y_size)

        return recon_x, mu, logvar
    
    def encode(self, x):
        org_size = x.size()
        y_size = (org_size[0], org_size[1], org_size[2] // 3)
        batch = org_size[0]
        x = x.view(batch, -1)

        h = self.encoder(x)
        return h
```

```python
recon_loss = nn.MSELoss(reduction='none')
kl_loss = lambda mu, logvar: -0.5 * torch.sum(1 + logvar - mu.pow(2) - logvar.exp())
# recon_loss = nn.CrossEntropyLoss()
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

model = ResAE(INPUT_DIM, INTER_DIM_1, INTER_DIM_2, INTER_DIM_3, LATENT_DIM, OUTPUT_DIM)
model.to(device)
optimizer = optim.Adam(model.parameters(), lr=1e-3)
# scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=3, gamma=0.1)

# Define the file path and name
model_info_file = os.path.join(model_save_dir, 'model.txt')

# Open the file in write mode
with open(model_info_file, 'w') as f:
    f.write(str(model))
    f.write('\n\n')
    f.write(str(summary(model, input_size=(BATCH_SIZE, FRAMES_IN_SEGMENT, MFCC_DIM * 3))))
model
```

```python
def draw_learning_curve(losses, others, epoch="", save_path=None): 
    train_losses, valid_losses, train_recons, valid_recons, train_kls, valid_kls= losses
    best_val_loss = others
    plt.clf()
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 4))
    ax1.plot(train_losses, label='Train')
    ax1.plot(valid_losses, label='Valid')
    ax1.axvline(x=best_val_loss, color='r', linestyle='--', label=f'Best: {best_val_loss}')
    ax1.set_title("Learning Curve Loss" + f" {epoch}")
    ax1.legend(loc="upper right")

    ax2.plot(train_recons, label='Train')
    ax2.plot(valid_recons, label='Valid')
    ax2.set_title('Learning Curve Recon' + f" {epoch}")
    ax2.legend(loc="upper right")

    ax3.plot(train_kls, label='Train')
    ax3.plot(valid_kls, label='Valid')
    ax3.set_title('Learning Curve KL' + f" {epoch}")
    ax3.legend(loc="upper right")

    plt.tight_layout()
    plt.xlabel("Epoch")
    display.clear_output(wait=True)
    display.display(plt.gcf())
    if save_path: 
        plt.savefig(save_path)
```

```python
def train(best_valid_loss, best_valid_loss_epoch): 
    for epoch in range(BASE, BASE + EPOCHS):
        model.train()
        train_loss = 0.
        train_recon = 0.
        train_kl = 0.
        train_num = len(train_loader)    # train_loader
        for idx, (x, y) in enumerate(train_loader):
            optimizer.zero_grad()
            batch = x.size(0)
            x = x.to(device)
            y = y.to(device)

            x_hat, mu, logvar = model(x)
            # print(x_hat.size(), y.size())
            # loss = recon_loss(x_hat, y)
            recon = recon_loss(x_hat, y)
            recon = recon.view(size=(batch, -1)).mean(dim=-1).sum()
            kl = kl_loss(mu, logvar)
            loss = (recon + kl) / batch
            train_loss += loss.item()
            train_recon += recon.item()
            train_kl += kl.item()
            
            loss.backward()
            
            # torch.nn.utils.clip_grad_norm_(model.parameters(), clip_value)
            # 这个函数计算的是全局梯度范数
            # torch.nn.utils.clip_grad_norm(parameters=model.parameters(), max_norm=5, norm_type=2)
            # torch.nn.utils.clip_grad_norm_(parameters=model.parameters(), max_norm=5, norm_type=2)
            # parameters: an iterable of Variables that will have gradients normalized
            # max_norm: max norm of the gradients(阈值设定)
            # norm_type: type of the used p-norm. Can be'inf'for infinity norm(定义范数类型)

            optimizer.step()
            # scheduler.step()

            draw_progress_bar(idx, train_num)

        train_losses.append(train_loss / train_num)
        train_recons.append(train_recon / train_num)
        train_kls.append(train_kl / train_num)

        last_model_name = f"{epoch}.pt"
        torch.save(model.state_dict(), os.path.join(model_save_dir, last_model_name))

        model.eval()
        valid_loss = 0.
        valid_recon = 0.
        valid_kl = 0.
        valid_num = len(valid_loader)
        for idx, (x, y) in enumerate(valid_loader):
            batch = x.size(0)
            x = x.to(device)
            y = y.to(device)

            x_hat, mu, logvar = model(x)
            recon = recon_loss(x_hat, y)
            recon = recon.view(size=(batch, -1)).mean(dim=-1).sum()
            kl = kl_loss(mu, logvar)
            loss = (recon + kl) / batch
            # loss = recon_loss(x_hat, y)
            
            valid_loss += loss.item()
            valid_recon += recon.item()
            valid_kl += kl.item()

            # draw_progress_bar(idx, valid_num)

        avg_valid_loss = valid_loss / valid_num
        valid_losses.append(avg_valid_loss)
        valid_recons.append(valid_recon / valid_num)
        valid_kls.append(train_kl / train_num)
        if avg_valid_loss < best_valid_loss: 
            best_valid_loss = avg_valid_loss
            best_valid_loss_epoch = epoch

        draw_learning_curve(losses=(train_losses.get(), valid_losses.get(), 
                                    train_recons.get(), valid_recons.get(), 
                                    train_kls.get(), valid_kls.get()),
                            others=best_valid_loss_epoch, 
                            epoch=str(epoch))
        train_losses.save()
        valid_losses.save()

        train_recons.save()
        valid_recons.save()

        train_kls.save()
        valid_kls.save()

    return best_valid_loss, best_valid_loss_epoch
```
