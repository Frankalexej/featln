```python
import matplotlib.pyplot as plt
import math

def calculate_in_group_pairs(group_size):
    return 2 * math.comb(group_size, 2)

def calculate_between_group_pairs(group_size):
    return group_size * group_size

def plot_pairs_vs_group_size(group_sizes):
    in_group_pairs = [calculate_in_group_pairs(size) for size in group_sizes]
    between_group_pairs = [calculate_between_group_pairs(size) for size in group_sizes]

    plt.plot(group_sizes, in_group_pairs, label='In-group Pairs')
    plt.plot(group_sizes, between_group_pairs, label='Between-group Pairs')

    plt.xlabel('Group Size')
    plt.ylabel('Number of Pairs')
    plt.title('In-group Pairs vs Between-group Pairs')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example: Group sizes from 1 to 50
group_sizes = list(range(1, 1000, 10))
plot_pairs_vs_group_size(group_sizes)
```

```python
def separate_dataframe(df, n):
    # Convert the dataframe to a numpy array
    array = df.to_numpy()

    # Split the array into n parts
    parts = np.array_split(array, n)

    # Convert each part back to a dataframe
    dfs = [pd.DataFrame(part, columns=df.columns) for part in parts]

    return dfs
```

```python
def separate_list(lst, n):
    sublists = [[] for _ in range(n)]
    index = 0
    for item in lst:
        sublists[index].append(item)
        index = (index + 1) % n
    return sublists
```

```python
def split_list(input_list, n):
    # Calculate the length of each sublist
    sublist_length = len(input_list) // n
    
    # Calculate the remainder for handling uneven division
    remainder = len(input_list) % n
    
    # Initialize the starting index for slicing
    start = 0
    
    # Iterate through the number of sublists
    for i in range(n):
        # Calculate the ending index for slicing
        end = start + sublist_length + (1 if i < remainder else 0)
        
        # Yield the sublist
        yield input_list[start:end]
        
        # Update the starting index for the next iteration
        start = end

# Example usage:
original_list = list(range(10000))
num_sublists = 10

result = list(split_list(original_list, num_sublists))
print(result)
```

```python
import random
import string

def generate_random_string(length):
    letters = string.ascii_lowercase
    return ''.join(random.choice(letters) for _ in range(length))

def generate_random_list(num_strings, string_length):
    random_list = []
    for _ in range(num_strings):
        random_string = generate_random_string(string_length)
        random_list.append(random_string)
    return random_list

# Example usage: Generate a random list with 5 strings of length 8
random_list = generate_random_list(5, 8)
print(random_list)
```

```python
def norm_mvn(mel_spec):
    eps = 1e-9
    mean = mel_spec.mean()
    std = mel_spec.std()
    norm_spec = (mel_spec - mean) / (std + eps)
    return norm_spec
```
